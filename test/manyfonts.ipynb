```python
import urllib.request
import os

import ziafont as zf
```

```python
def testfont(url):
    name = os.path.split(url)[1]
    if not os.path.exists(name):
        urllib.request.urlretrieve(str(url), name)
    font = zf.Font(name)
    if 'latn' in font.scripts():
        font.language('latn', '')    
    print(f"Font: {font.info.names.name}; Type: {'CFF' if 'CFF ' in font.tables else 'glyf'}")
    display(font.text('ABCDEFGHIJKLMNOPQRSTUVWXYZ'))
    display(font.text('ABCDEFGHIJKLMNOPQRSTUVWXYZ'.lower()))
    display(font.text('0123456789!@#$%^&*()ÑÅÊò£<>'))
    display(font.text('αβγδεζηθικλμνξοπρστυφχψω'))
```
