```python
import xml.etree.ElementTree as ET
from IPython.display import SVG
```

```python
# Adding to existing SVG
svg = ET.Element('svg')
svg.attrib['width'] = '200'
svg.attrib['height'] = '220'
svg.attrib['xmlns'] = 'http://www.w3.org/2000/svg'
svg.attrib['viewBox'] = f'0 0 200 220'

font.str2svg('ABCD\n123', xy=(50, 20), canvas=svg, fontsize=18, halign='right', valign='base')
font.str2svg('ABCD\n123', xy=(100, 20), canvas=svg, fontsize=18, halign='center', valign='base')
font.str2svg('ABCD\n123', xy=(150, 20), canvas=svg, fontsize=18, halign='left', valign='base')

font.str2svg('ABCD\n123', xy=(50, 80), canvas=svg, fontsize=18, halign='right', valign='center')
font.str2svg('ABCD\n123', xy=(100, 80), canvas=svg, fontsize=18, halign='center', valign='center')
font.str2svg('ABCD\n123', xy=(150, 80), canvas=svg, fontsize=18, halign='left', valign='center')

font.str2svg('ABCD\n123', xy=(50, 110), canvas=svg, fontsize=18, halign='right', valign='top')
font.str2svg('ABCD\n123', xy=(100, 110), canvas=svg, fontsize=18, halign='center', valign='top')
font.str2svg('ABCD\n123', xy=(150, 110), canvas=svg, fontsize=18, halign='left', valign='top')

font.str2svg('ABCD\n123', xy=(50, 210), canvas=svg, fontsize=18, halign='right', valign='bottom')
font.str2svg('ABCD\n123', xy=(100, 210), canvas=svg, fontsize=18, halign='center', valign='bottom')
font.str2svg('ABCD\n123', xy=(150, 210), canvas=svg, fontsize=18, halign='left', valign='bottom')

SVG(ET.tostring(svg))
```

```python
assert 'symbol' in font.glyph('a').svg()
```

```python
# SVG 1.x mode
font.svg2 = False

svg = ET.Element('svg')
svg.attrib['width'] = '200'
svg.attrib['height'] = '220'
svg.attrib['xmlns'] = 'http://www.w3.org/2000/svg'
svg.attrib['viewBox'] = f'0 0 200 220'

font.str2svg('ABCD\n123', xy=(50, 20), canvas=svg, fontsize=18, halign='right', valign='base')
font.str2svg('ABCD\n123', xy=(100, 20), canvas=svg, fontsize=18, halign='center', valign='base')
font.str2svg('ABCD\n123', xy=(150, 20), canvas=svg, fontsize=18, halign='left', valign='base')

font.str2svg('ABCD\n123', xy=(50, 80), canvas=svg, fontsize=18, halign='right', valign='center')
font.str2svg('ABCD\n123', xy=(100, 80), canvas=svg, fontsize=18, halign='center', valign='center')
font.str2svg('ABCD\n123', xy=(150, 80), canvas=svg, fontsize=18, halign='left', valign='center')

font.str2svg('ABCD\n123', xy=(50, 110), canvas=svg, fontsize=18, halign='right', valign='top')
font.str2svg('ABCD\n123', xy=(100, 110), canvas=svg, fontsize=18, halign='center', valign='top')
font.str2svg('ABCD\n123', xy=(150, 110), canvas=svg, fontsize=18, halign='left', valign='top')

font.str2svg('ABCD\n123', xy=(50, 210), canvas=svg, fontsize=18, halign='right', valign='bottom')
font.str2svg('ABCD\n123', xy=(100, 210), canvas=svg, fontsize=18, halign='center', valign='bottom')
font.str2svg('ABCD\n123', xy=(150, 210), canvas=svg, fontsize=18, halign='left', valign='bottom')

assert 'symbol' not in ET.tostring(svg, encoding='unicode')
SVG(ET.tostring(svg))
```
