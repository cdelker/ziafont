```python
import xml.etree.ElementTree as ET
from IPython.display import SVG
```

```python
svg = ET.Element('svg')
svg.attrib['width'] = '200'
svg.attrib['height'] = '200'
svg.attrib['xmlns'] = 'http://www.w3.org/2000/svg'
svg.attrib['viewBox'] = f'0 0 200 200'

rect = ET.SubElement(svg, 'rect')
rect.attrib['x'] = '0'
rect.attrib['y'] = '0'
rect.attrib['width'] = '200'
rect.attrib['height'] = '200'
rect.attrib['fill'] = 'none'
rect.attrib['stroke'] = 'red'

font.str2svg('ABCD\n123', xy=(50, 50), canvas=svg, fontsize=18, halign='center')
font.str2svg('Ziafont\nhey!', xy=(100, 150), canvas=svg, linespacing=.7, fontsize=18, halign='center')

SVG(ET.tostring(svg))
```
