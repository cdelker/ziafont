```python
import xml.etree.ElementTree as ET
from IPython.display import SVG
```

```python
# Adding to existing SVG
svg = ET.Element('svg')
svg.attrib['width'] = '200'
svg.attrib['height'] = '220'
svg.attrib['xmlns'] = 'http://www.w3.org/2000/svg'
svg.attrib['viewBox'] = f'0 0 200 220'

font.text('ABCD\n123', size=18, halign='right', valign='base').drawon(svg, 50, 20)
font.text('ABCD\n123', size=18, halign='center', valign='base').drawon(svg, 100, 20)
font.text('ABCD\n123', size=18, halign='left', valign='base').drawon(svg, 150, 20)

font.text('ABCD\n123', size=18, halign='right', valign='center').drawon(svg, 50, 80)
font.text('ABCD\n123', size=18, halign='center', valign='center').drawon(svg, 100, 80)
font.text('ABCD\n123', size=18, halign='left', valign='center').drawon(svg, 150, 80)

font.text('ABCD\n123', size=18, halign='right', valign='top').drawon(svg, 50, 110)
font.text('ABCD\n123', size=18, halign='center', valign='top').drawon(svg, 100, 110)
font.text('ABCD\n123', size=18, halign='left', valign='top').drawon(svg, 150, 110)

font.text('ABCD\n123', size=18, halign='right', valign='bottom').drawon(svg, 50, 210)
font.text('ABCD\n123', size=18, halign='center', valign='bottom').drawon(svg, 100, 210)
font.text('ABCD\n123', size=18, halign='left', valign='bottom').drawon(svg, 150, 210)

SVG(ET.tostring(svg))
```

```python
# SVG 1.x mode
zf.config.svg2 = False

svg = ET.Element('svg')
svg.attrib['width'] = '200'
svg.attrib['height'] = '220'
svg.attrib['xmlns'] = 'http://www.w3.org/2000/svg'
svg.attrib['viewBox'] = f'0 0 200 220'

font.text('ABCD\n123', size=18, halign='right', valign='base').drawon(svg, 50, 20)
font.text('ABCD\n123', size=18, halign='center', valign='base').drawon(svg, 100, 20)
font.text('ABCD\n123', size=18, halign='left', valign='base').drawon(svg, 150, 20)

font.text('ABCD\n123', size=18, halign='right', valign='center').drawon(svg, 50, 80)
font.text('ABCD\n123', size=18, halign='center', valign='center').drawon(svg, 100, 80)
font.text('ABCD\n123', size=18, halign='left', valign='center').drawon(svg, 150, 80)

font.text('ABCD\n123', size=18, halign='right', valign='top').drawon(svg, 50, 110)
font.text('ABCD\n123', size=18, halign='center', valign='top').drawon(svg, 100, 110)
font.text('ABCD\n123', size=18, halign='left', valign='top').drawon(svg, 150, 110)

font.text('ABCD\n123', size=18, halign='right', valign='bottom').drawon(svg, 50, 210)
font.text('ABCD\n123', size=18, halign='center', valign='bottom').drawon(svg, 100, 210)
font.text('ABCD\n123', size=18, halign='left', valign='bottom').drawon(svg, 150, 210)

assert 'symbol' not in ET.tostring(svg, encoding='unicode')
SVG(ET.tostring(svg))
```

```python
# Rotations - default mode
# (See https://matplotlib.org/stable/gallery/text_labels_and_annotations/demo_text_rotation_mode.html)
svg = ET.Element('svg')
svg.attrib['xmlns'] = 'http://www.w3.org/2000/svg'
svg.attrib['xmlns:xlink'] = 'http://www.w3.org/1999/xlink'
svg.attrib['width'] = '500'
svg.attrib['height'] = '300'
svg.attrib['viewBox'] = '0 0 500 300'

for xy, halign, valign in [((20, 20), 'left', 'top'),
                           ((100, 20), 'center', 'top'),
                           ((180, 20), 'right', 'top'),
                           ((20, 80), 'left', 'center'),
                           ((100, 80), 'center', 'center'),
                           ((180, 80), 'right', 'center'),
                           ((20, 130), 'left', 'base'),
                           ((100, 120), 'center', 'base'),
                           ((180, 110), 'right', 'base'),
                           ((20, 180), 'left', 'bottom'),
                           ((100, 180), 'center', 'bottom'),
                           ((180, 180), 'right', 'bottom'),

                          ]:
    circ = ET.SubElement(svg, 'circle')
    circ.set('cx', str(xy[0]))
    circ.set('cy', str(xy[1]))
    circ.set('r', '3')
    circ.set('stroke', 'blue')
    circ.set('fill', 'blue')

    font.text('Hello\n123', rotation=30,
                 size=12,
                 rotation_mode = 'default',
                 halign=halign, valign=valign).drawon(svg, xy[0], xy[1])

SVG(ET.tostring(svg))
```

```python
# Rotations - anchor mode
# (See https://matplotlib.org/stable/gallery/text_labels_and_annotations/demo_text_rotation_mode.html)
svg = ET.Element('svg')
svg.attrib['xmlns'] = 'http://www.w3.org/2000/svg'
svg.attrib['xmlns:xlink'] = 'http://www.w3.org/1999/xlink'
svg.attrib['width'] = '500'
svg.attrib['height'] = '300'
svg.attrib['viewBox'] = '0 0 500 300'

for xy, halign, valign in [((20, 20), 'left', 'top'),
                           ((100, 20), 'center', 'top'),
                           ((180, 20), 'right', 'top'),
                           ((20, 80), 'left', 'center'),
                           ((100, 80), 'center', 'center'),
                           ((180, 80), 'right', 'center'),
                           ((20, 130), 'left', 'base'),
                           ((100, 120), 'center', 'base'),
                           ((180, 120), 'right', 'base'),
                           ((20, 180), 'left', 'bottom'),
                           ((100, 180), 'center', 'bottom'),
                           ((180, 180), 'right', 'bottom'),

                          ]:
    circ = ET.SubElement(svg, 'circle')
    circ.set('cx', str(xy[0]))
    circ.set('cy', str(xy[1]))
    circ.set('r', '4')
    circ.set('stroke', 'blue')
    circ.set('fill', 'blue')

    font.text('Hello\n123', rotation=30,
                 size=12,
                 rotation_mode = 'anchor',
                 halign=halign, valign=valign).drawon(svg, xy[0], xy[1])

SVG(ET.tostring(svg))
```
